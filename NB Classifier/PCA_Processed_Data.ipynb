```python
import pandas as pd
```

```python
df = pd.read_csv('X_res.csv')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(df)
```

```python
pca
```

```python
data2D = pca.transform(df)
```

```python
%matplotlib inline
```

```python
df = df.rename(columns = {df.columns[0] : 'Index'})
```

```python
#PLOTTING DATA USING PCA
import matplotlib.pyplot as plt
plt.figure()
plt.legend([df.Index])
plt.scatter(data2D[:,0], data2D[:,1], c=df['Index'],cmap=plt.cm.get_cmap('viridis', 10), edgecolor='none', alpha=0.5)
plt.colorbar()
plt.show() 
```

```python
#As seen below, the amount of variance for number of components for 2 is more than 99% Hence PCA has less loss of information
import numpy as np
pca = PCA().fit(df)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```
